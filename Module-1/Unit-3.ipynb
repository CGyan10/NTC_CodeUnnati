```python
import numpy as np
x = np.array([2,3,8,7])
y = np.array([1,4,1,9])

plt.plot(x, y, marker = '*', c = 'c', linestyle = 'dotted')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Line Plot")
plt.show()
```

```python
#import matplotlib.pyplot as plt
import random as ran

students = ["Jane","Joe","Beck","Tom",
			"Sam","Eva","Samuel","Jack"]

marks=[]
for i in range(0,len(students)):
	marks.append(ran.randint(0, 101))

plt.xlabel("Students")
plt.xticks(rotation = 45)
plt.ylabel("Marks")
plt.yticks(rotation = 45)
plt.title("CLASS RECORDS")
plt.plot(students, marks, color = 'green',
         linestyle = 'solid', marker = 'o',
         markerfacecolor = 'red', markersize = 12)
```

# Adding Annotatae & legend

```python
import numpy as np
import matplotlib.pyplot as plt
#you cal also write    
#from matplotlib import pyplot as plt


x = np.arange(0,10)
y = x^2
z=  x^3

plt.plot(x,y)
plt.plot(y,z)
plt.plot(x,z)

plt.title(" Graph Drawing")
plt.xlabel("Time")
plt.ylabel("Distance")


#Annotataion 

plt.annotate(xy=[2,1], text="First Entry", rotation= 20)

#add legend

plt.legend(["Race 1", "Race2", "Race3"], loc = 4)
plt.show()
```

# Pie chart

pie chart is a circular stastical plot that can display only one series of data. 

the area of the chart is the total % of the given data

```python
#import librarties

import numpy as np
import matplotlib.pyplot as plt

#creating datset
Branch = ["Computer","Electrical", "Mechnical", "Civil", "Automobile","IT"]

data = [50, 34, 40, 29, 12, 23]
explode = (0.1,0,0,0.1,0,0)
#create chart

fig = plt.figure(figsize=(10,7))
plt.pie(data, labels=Branch, explode=explode, autopct='%1.1f%%', shadow = True)

plt.show()
```

# 3D ploting

```python
#import mplot3d toolkit, numpy & matplotlib

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d #3d toolkit lib

```

```python
fig = plt.figure()

#syntax for 3D projection

ax = plt.axes(projection = '3d')

#defining all 3 axes

z= np.linspace(0,1,100)
x = z * np.sin(25*z)
y = z * np.cos(25*z)

#3d Ploting

ax.plot3D(x, y, z, 'green')
ax.set_title('3D ploting')
plt.show()
```

# Saving the chart
